## 1. In Which year the percentage of air pollution is high?

```python
city_day_date.groupby('year').size().plot(kind='pie', figsize=(6, 6), autopct='%1.2f%%')
plt.ylabel("")
plt.title("Percentage of Air pollution from the year 2015 to 2020")
## Reference[3]
```

> In this pie chart, 2019 at peak point in the proportion of air pollution, which was 30.50% and represents by purple colour. The rate of air pollution was quite less in 2015 and 2016,that was 5.04%-6.70%. Ratio of air pollution was mostly equal in the years of 2018 and 2020, it was approximately nearby a quarter whereas in 2017, this percentage was only 11.34%.


## 2. In which season the proportion of air pollution is less?

```python
city_day_trans.groupby('Season').size().plot(kind='pie',figsize=(6, 6),autopct='%1.1f%%')
plt.ylabel("")
plt.title("The percentage of air pollution according to season")
## Reference[3]
```

> It is noticed that, in winter, air pollution is very less in comparison to other seasons, that is only 17.4% while the ratio of air pollution in monsoon at highest point, this is about 15% more than winter. Moreover, post-monsoon and summer seasons are both approximately lies in 23.1%-27.4% in the percentage of air pollution.

## 3. Grouping the CO by city and calculating the average CO per city?

```python
x=pd.DataFrame(AQI_data.groupby(['City'])[['CarbonDioxide']].mean().sort_values(by='CarbonDioxide',ascending=False).head())
x=x.reset_index('City')
plt.figure(figsize=(12,12))
sns.relplot(data=x,x='CarbonDioxide',y='City', hue='City')
plt.title("The average of CO as per city")
## Reference[1]
```

> The graph depicts the average CO per city, with Delhi coming in first with a value of around 2.25 and a dark blue colour. 
Furthermore, CO levels in Patna are less than 1.25, whereas CO levels in Gurugram and Amritsar are only 0.75, and CO levels in 
Visakhapatnam are slightly higher than in both cities

## 4. In which city the average of NO gas is higher or lesser?

```python
x=pd.DataFrame(AQI_data.groupby(['City'])[['NitricOxide']].mean().sort_values(by='NitricOxide',ascending=False).head())
x=x.reset_index('City')
plt.figure(figsize=(8,8))
sns.barplot(data=x,x='City',y='NitricOxide')
plt.title("The average of NitrcOxide as per city")
## Reference[1]
```

> This graph shows that amount of NO gas in patna city is at the peak which is more than 50 whereas vishakhapatnam is very low
which is near about 15. After that the ratio of NO gas in kolkata and amritsar are approximately similar which is about 20. 
In the case of delhi level of NO gas is 33.

## 5. In which city the average of NO2 gas is higher or lesser?

```python
x=pd.DataFrame(AQI_data.groupby(['City'])[['NitrogenDioxide']].mean().sort_values(by='NitrogenDioxide',ascending=False).head())
x=x.reset_index('City')
plt.figure(figsize=(8,8))
sns.barplot(data=x,x='City',y='NitrogenDioxide')
plt.title("The average of Nitrogen Dioxide (NO2) as per city")
## Reference[1]
```

> In the vertical bar graph, Delhi is ranked at the peak in the average of NO2, which is represented by the color blue. 
Furthermore,Visakhapatnam is the second-most populous city in India (NO2). Patna and Hyderabad are ranked fourth and fifth, 
respectively,in terms of CO average, but Kolkata is quite low as comparison to other cities.
